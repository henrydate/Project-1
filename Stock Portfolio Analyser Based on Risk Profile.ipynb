```python
# Initial imports
import pandas as pd
import hvplot.pandas
import yfinance as yf
from MCForecastTools import MCSimulation
%matplotlib inline
```

```python
#Here we are fetching all the initial ticker data 
# Fetch high growth ticker data
aapl = yf.Ticker("AAPL")
tsla = yf.Ticker("TSLA")
nke = yf.Ticker("NKE")
amzn = yf.Ticker("AMZN")
wmt = yf.Ticker("WMT")
# Fetch low volatility ticker data
ldos = yf.Ticker("LDOS")
incy = yf.Ticker("INCY")
nbix = yf.Ticker("NBIX")
ko = yf.Ticker("KO")
aiz = yf.Ticker("AIZ")
```

```python
# get historical market data for high growth ticker data
hist_appl = aapl.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
hist_tsla = tsla.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
hist_nke = nke.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
hist_amzn = amzn.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
hist_wmt = wmt.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
# get historical market data for low volatility ticker data
hist_ldos = ldos.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
hist_incy = incy.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
hist_nbix = nbix.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
hist_ko = ko.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
hist_aiz = aiz.history(period="5y").drop(columns=["Dividends", "Stock Splits"])
# Separate ticker data
```

```python
#Joining all historical data
his_stock_data = pd.concat([hist_appl, hist_tsla, hist_nke, hist_amzn, hist_wmt, hist_ldos, hist_incy, hist_nbix, hist_ko, hist_aiz], axis = 1, 
                           join="inner", keys=["appl", "tsla", "nke", "amzn", "wmt", "idos", "incy", "nbix", "ko", "aiz"])
```

```python
#testing the data to see if it looks pretty
his_stock_data
```

```python
MC_stock_data = MCSimulation(
    portfolio_data = his_stock_data,
    weights = [.10, .10, .10, .10, .10, .10, .10, .10, .10, .10, .10],
    num_simulation = 500,
    num_trading_days = 252 * 10
)
```
